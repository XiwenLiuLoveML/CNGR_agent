# Lab 1：第一次调用 LLM

今天这一段，我们先做三件事：

1. 接上模型
2. 看懂最原始的 messages
3. 改几次 prompt，观察结果变化

````python
# 导入工具
from openai import OpenAI
from dotenv import load_dotenv
import os
from IPython.display import display, Markdown

# 读取 .env 里的 API key
load_dotenv()

# 创建客户端（通义千问兼容 OpenAI 接口）
client = OpenAI(
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
    api_key=os.getenv("API_KEY")
)
````

````python
# 构造对话消息
messages = [
    {"role": "system", "content": "你是一个严谨、专业的咨询顾问，说话非常正式"},
    {"role": "user", "content": "请介绍一下你自己"}
]

response = client.chat.completions.create(
    model="qwen3.6-plus",
    messages=messages
)

display(Markdown(response.choices[0].message.content))
````

## 什么是 messages？

这里面最重要的有两个角色：

- `system`：设定 AI 是谁、用什么风格说话
- `user`：告诉 AI 你现在要它做什么

你可以先把它理解成：

- `system` = 人设
- `user` = 问题

````python
#尝试不同的系统提示语，看看对话风格有什么不同
messages = [
    {"role": "system", "content": "你是一个轻松、幽默的职业顾问，说话自然一点"},
    {"role": "user", "content": "请介绍一下你自己"}
]

response = client.chat.completions.create(
    model="qwen3.6-plus",
    messages=messages
)

display(Markdown(response.choices[0].message.content))
````

````python
# 再试试其他的提示语，比如让它扮演一个职业数字人，看看它会怎么介绍自己
messages = [
    {"role": "system", "content": "你是一个职业数字人，正在帮助用户介绍自己的工作背景和职业方向"},
    {"role": "user", "content": "请介绍一下你自己"}
]

response = client.chat.completions.create(
    model="qwen3.6-plus",
    messages=messages
)

display(Markdown(response.choices[0].message.content))
````

## 你刚刚已经看到一件很重要的事

同一个问题，
只要 `system prompt` 变了，
输出风格就会明显变化。

这就是为什么 prompt 会影响结果。

````python
# 最后我们把调用模型的代码封装成一个函数，方便后续调用

def call_llm(user_question: str, system_prompt: str):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_question}
    ]

    response = client.chat.completions.create(
        model="qwen3.6-plus",
        messages=messages
    )

    content = response.choices[0].message.content

    # 👉 去掉可能的 Markdown 标题（避免字体变大）
    content = content.replace("###", "").replace("##", "").replace("#", "")

    display(Markdown(f"""
    ### 🤖 AI回答

    ---
    {content}
    ---
    """))
````

````python
# 调用函数试试
system_prompt = "你是一个积极乐观的职业数字人，帮助用户介绍自己的工作背景和职业方向"
question = "请介绍一下你自己"

answer = call_llm(question, system_prompt)
answer
````

## 🧠 让 AI 了解你，而不是“假装你”

所以接下来，我们换一种方式。

👉 **把你的资料给它**

## 🚀 接下来我们要做什么？

````python
# 读取简历和工作日志文件
with open("data/CV.txt", "r", encoding="utf-8") as f:
    resume = f.read()

with open("data/work_log.txt", "r", encoding="utf-8") as f:
    work_log = f.read()
````

到这里，我们做一件很关键的事情。

👉 不再让 AI“设定人设”，  
👉 而是让它基于你的资料来说话。

---

下面这个 prompt 很重要，
你可以先不用完全理解，
先运行一遍，看看效果。

````python
system_prompt = """
你不是在扮演这个人。

你是在“基于他的资料”，替他说话。

请遵循以下规则：
- 使用第一人称（我）
- 只基于提供的资料，不要编造不存在的经历
- 语气自然、真实，有一点思考感
- 不要像简历，不要像机器人
"""

user_question = f"""
这是我的职业背景（CV）：
{resume}

这是我最近几天的工作记录：
{work_log}

请用第一人称介绍我是谁，我在做什么，
可以带一点我最近在思考的问题。
"""
````

````python
# 调用 LLM 函数，看看它怎么介绍这个人
call_llm(user_question, system_prompt)
````
